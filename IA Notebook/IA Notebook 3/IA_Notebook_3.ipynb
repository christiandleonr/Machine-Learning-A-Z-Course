```python
# Gradient Descent

# Importing the libraries 

import numpy as np
import matplotlib.pyplot as plt
```

```python
# lambda is to create a anonymous function, x = th[0] and y = th [1]

func = lambda th: np.sin(1/2 * th[0] ** 2 - 1/4 * th[1] ** 2 + 3) * np.cos(2 * th[0] + 1 - np.e ** th[1])
```

```python
# Visualising the slope Map

res = 100

_X = np.linspace(-2, 2, res)
_Y = np.linspace(-2, 2, res)

_z = np.zeros((res, res))

# enumerate return the index and the value ix = index, x = value

for ix, x in enumerate(_X):
    for iy, y in enumerate(_Y):
        # here we have to put first the index iy because iy represent y axis which the same as rows
        _z[iy, ix] = func([x, y])

plt.contourf(_X, _Y, _z, 100)
plt.colorbar()

# Generating a random point
"""We have to generate a random point to start with the gradient descent method"""

theta = np.random.rand(2) * 4 - 2

_T = np.copy(theta)

h = 0.0001
lr = 0.0001

plt.plot(theta[0], theta[1], "*", c='white')

grad = np.zeros(2)

for _ in range(10000):
    for it, th in enumerate(theta):
        _T = np.copy(theta)
        
        _T[it] = _T[it] + h
        
        derived = (func(_T) - func(theta)) / h
        
        grad[it] = derived
        
    theta = theta - lr * grad
    
    if _ % 100 == 0:
        plt.plot(theta[0], theta[1], ".", c='red')

plt.plot(theta[0], theta[1], "*", c='cyan')
plt.show()
```
