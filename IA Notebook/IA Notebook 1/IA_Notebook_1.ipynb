```python
# Linear Regression from Scratch

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
print(boston.DESCR)
```

Minimize the mean square error (MCO): $\beta = (X^{T}X)^{-1}X^{T}Y$

```python
X = np.array(boston.data[:, 5])
y = np.array(boston.target)

plt.scatter(X, y, alpha=0.3)
plt.xlabel('Media of rooms')
plt.ylabel('Value of the house')
plt.title('Price of house in boston')
```

```python
# Add ones column that represent the independent value
X = np.array([np.ones(506), X]).T

"""
np.linalg.inv is used to obtain the inverse of a value
@ is used to obtain matrix multiplication
.T is used to obtain the transposed of a matrix
"""
B = np.linalg.inv(X.T @ X) @ X.T @ y

plt.plot([4, 9], [B[0]+B[1]*4, B[0]+B[1]*9], c='red')
plt.show()
```
